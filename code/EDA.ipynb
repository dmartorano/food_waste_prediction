```python
columns_to_drop = ["sub_sector", "sub_sector_category", "food_category", "upstream_mtco2e_footprint", "downstream_mtco2e_footprint"]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

The first three dropped columns are logically the least directly-related categories to the amount of CO2 emission, while the upstream and downstream MT CO2e footprints are aggregated in another value. As for the other two categorical variables, sector would determine possible spoilage practices and food_type would be potentially indicative of shelf life.

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['sector', 'food_type'], dtype=int)
```

Use get_dummies to turn sector and food_type into usable numerical features that will be helpful for determining how different sectors deal with food surplus and which kinds of surplus food are disposed of in what ways.

```python
df["tons_productive_surplus"] = df[["tons_donated", "tons_composted", "tons_anaerobically_digested", "tons_land_application", "tons_biomaterial_processing", "tons_animal_feed"]].sum(axis=1)
```

Creates an aggregate column for all ways of processing surplus food that would be considered productive rather than simply disposing of the food.

```python
df["tons_unproductive_surplus"] = df[["tons_not_harvested", "tons_incinerated", "tons_landfilled", "tons_sewer", "tons_refuse_discards"]].sum(axis=1)
```

Creates an aggregate column for all ways of processing surplus food that would not be considered productive.

```python
df["tons_consumed"] = df["tons_supply"] - df["tons_surplus"]
```

Creates a column which represents the amount of food that is distributed and consumed through standard means, i.e. not excess.

```python
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
df.dtypes
```

```python
df.to_csv("../data/clean_us_food_waste.csv", index=False)
```

## Exploratory Data Analysis

#### Supplementary DataFrames

```python
df_true_numeric = df[["tons_consumed", "tons_surplus", "tons_supply", "us_dollars_surplus", "tons_waste", "tons_uneaten", "tons_inedible_parts", 
                      "tons_not_fit_for_human_consumption", "tons_donated", "tons_biomaterial_processing", "tons_animal_feed", "tons_anaerobically_digested", 
                      "tons_composted", "tons_not_harvested", "tons_incinerated", "tons_land_application", "tons_landfilled", "tons_sewer", "tons_refuse_discards", 
                      "total_mtco2e_footprint"]]
```

```python
df_time = df.groupby("year").sum()
```

```python
df_time_features = df_time[["tons_consumed", "tons_donated", "tons_biomaterial_processing", "tons_animal_feed", "tons_anaerobically_digested", 
                            "tons_composted", "tons_incinerated", "tons_land_application", "tons_landfilled", "tons_sewer", "tons_refuse_discards"]]
```

These supplementary dataframes are specialized to focus on only numeric data (true numeric data, i.e. no booleans converted to integers), annual aggregational data, and annual aggregation data without redundancies, respectively. The numeric dataframe will be particularly helpful in determining correlational data, the annual data will provide a dataframe to use when observing data in specific years or over time, and the non-redundant annual data can be used to observe annual distribution data.



### Correlational Analysis

```python
plt.figure(figsize = (15,15))

corr = df_true_numeric.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, mask = mask, annot = True, fmt=".2f", vmin = -1, vmax = 1)
plt.savefig("../visualizations/heatmap.png");
```

The heatmap above shows us, in particular, how a lot of the more common modes of disposing of/mitigating food waste are the most closely correlated to the carbon footprint of the supply chain. It is also the case that these modes are likely the most impactful, causing some of the larger footprints.

### Time Series Analysis

```python
df_time
```

```python
def plot_series(df, cols=None, title='Title', xlab=None, ylab=None, steps=1):
    
    plt.figure(figsize=(18,9))
    
    for col in cols:
            
        plt.plot(df[col], label=col)
        
    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    
    plt.yticks(fontsize=18)
    plt.xticks(df.index[0::steps], fontsize=18)
    
    plt.legend();
```

```python
plot_series(df_time, [["tons_surplus", "tons_productive_surplus", "tons_unproductive_surplus"]], title="Generation of U.S. Food Waste from 2010 to 2022", ylab="Waste, in 10 million tons")
```
