```python
columns_to_drop = ["sub_sector", "sub_sector_category", "food_category"]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

The dropped columns are logically the least directly-related categories to the amount of food wasted, whereas sector would determine possible spoilage practices and food_type would be potentially indicative of shelf life.

```python
df = pd.get_dummies(df, columns=['sector', 'food_type'], dtype=int)
```

Use get_dummies to turn sector and food_type into usable numerical features

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```
